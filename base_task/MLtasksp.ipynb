```python
y=train_df["label"]
X=train_df.drop("label",axis=1)
```

```python
X=X.to_numpy()
y=y.to_numpy()
```

```python
X=X/255
```

creating the dataset and splitting them as training (80 %) and validation (20%) sets

```python
dataset=TensorDataset(X,y)
train_size=int(0.8*len(dataset))
val_size=len(dataset)-train_size
train_dataset,val_dataset=random_split(dataset,[train_size,val_size])
```

```python
train_loader = DataLoader(
    train_dataset,
    batch_size=64,
    shuffle=True
)
val_loader=DataLoader( val_dataset,batch_size=64)
```

```python
#define and create model
class FashionMNISTModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.input_layer = nn.Linear(784,16)

        self.left1 = nn.Linear(16,8)
        self.left2 = nn.Linear(8,8)

        self.right1 = nn.Linear(16,12)
        self.right2 = nn.Linear(12,8)

        self.output = nn.Linear(16,10)

        self.relu = nn.ReLU()
    def forward(self, x):
        x = self.relu(self.input_layer(x))
        #left branch
        left = self.relu(self.left1(x))
        skip=left
        left = self.relu(self.left2(left))
        left = left + skip
        #right branch
        right = self.relu(self.right1(x))
        right = self.relu(self.right2(right))
        merged=torch.cat([left,right],dim=1)
        output=self.output(merged)
        return output

model=FashionMNISTModel()
```
